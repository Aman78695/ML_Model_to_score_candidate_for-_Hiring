```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/Applications_for_Machine_Learning_internship_edited.xlsx - Sheet2.csv',header=0)
```

```python
## Cleaning and converting the 'Performance_UG' column values to a percentage by removing non-numeric characters and splitting by '/',
#then calculating the percentage by dividing the first value by the second value.
df['Performance_UG'] = df['Performance_UG'].fillna('').str.replace(r'[^0-9./]+', '', regex=True)
df['Performance_UG'] = df['Performance_UG'].str.split('/').apply(lambda x: float(x[0]) / float(x[1]) * 100 if len(x) == 2 else np.nan)

```

```python
## Cleaning and converting the 'Performance_12' column values to a percentage by removing non-numeric characters and splitting by '/',
#then calculating the percentage by dividing the first value by the second value.
df['Performance_12'	] = df['Performance_12'].fillna('').str.replace(r'[^0-9./]+', '', regex=True)
df['Performance_12'] = df['Performance_12'].str.split('/').apply(lambda x: float(x[0]) / float(x[1]) * 100 if len(x) == 2 else np.nan)
```

```python
## Cleaning and converting the 'Performance_10' column values to a percentage by removing non-numeric characters and splitting by '/',
#then calculating the percentage by dividing the first value by the second value.
df['Performance_10'	] = df['Performance_10'].fillna('').str.replace(r'[^0-9./]+', '', regex=True)
df['Performance_10'] = df['Performance_10'].str.split('/').apply(lambda x: float(x[0]) / float(x[1]) * 100 if len(x) == 2 else np.nan)
```

```python
#filling the nan values with 0
df['Performance_UG'].fillna(0,inplace=True)
df['Performance_PG'].fillna(0,inplace=True)
df['Performance_12'].fillna(0,inplace=True)
df['Performance_10'].fillna(0,inplace=True)
```

```python
#dropping the name columns as there is no values
df.drop(columns=['Name'],inplace=True)
```

```python
def hire_best_intern(df):
    weights = {
        'Python (out of 3)': 0.15,
        'Machine Learning (out of 3)': 0.15,
        'Natural Language Processing (NLP) (out of 3)': 0.1,
        'Deep Learning (out of 3)': 0.1,
        'Other skills': 0.1,
        'Performance_PG': 0.05,
        'Performance_UG': 0.05,
        'Performance_12': 0.05,
        'Performance_10': 0.05,
        'Are you available for 3 months': 0.15,
        'Degree': 0.1,
        'Stream': 0.15,
        'Current Year Of Graduation': 0.05
    }

    scores = []
    for _, row in df.iterrows():
        python_score = row['Python (out of 3)'] * weights['Python (out of 3)']
        ml_score = row['Machine Learning (out of 3)'] * weights['Machine Learning (out of 3)']
        nlp_score = row['Natural Language Processing (NLP) (out of 3)'] * weights['Natural Language Processing (NLP) (out of 3)']
        dl_score = row['Deep Learning (out of 3)'] * weights['Deep Learning (out of 3)']
        other_skills_score = 0
        if isinstance(row['Other skills'], str):
            other_skills_score = len(row['Other skills'].split(',')) * weights['Other skills']
        availability_score = 1 if row['Are you available for 3 months'] == 'Yes' else 0
        degree_score = 1 if row['Degree'] else 0

        stream_score = 0
        if isinstance(row['Stream'], str):
            # Assign stream scores based on priority
            if row['Stream'] in ['Data Science', 'Data Analytics', 'Machine Learning', 'Artificial Intelligence']:
                stream_score = 1.5 * weights['Stream']  # Give higher weightage to priority streams
            elif row['Stream'] == 'Data Science And Analytics':
                stream_score = 1.3 * weights['Stream']  # Assign higher weightage to 'Data Science And Analytics'
            elif row['Stream'] == 'Computer Science':
                stream_score = 1.2 * weights['Stream']  # Assign higher weightage to 'Computer Science'
            else:
                stream_score = 1 * weights['Stream']  # Assign default weightage for other streams

        graduation_score = 1 if row['Current Year Of Graduation'] > 0 else 0

        # Uncomment the following lines if 'Performance_PG' is present in the DataFrame
        # pg_score = row['Performance_PG'] * weights['Performance_PG']

        # Calculate the percentage score for 'Performance_UG' and apply weight
        ug_score = row['Performance_UG'] * weights['Performance_UG']

        # Calculate the percentage scores for 'Performance_12' and 'Performance_10' and apply weights
        score_12 = row['Performance_12'] * weights['Performance_12']
        score_10 = row['Performance_10'] * weights['Performance_10']

        overall_score = (
            python_score +
            ml_score +
            nlp_score +
            dl_score +
            other_skills_score +
            availability_score +
            degree_score +
            stream_score +
            graduation_score +
            ug_score +
            score_12 +
            score_10
        )

        scores.append(overall_score)

    available_interns = df[df['Are you available for 3 months'] == 'Yes']

    sorted_interns = available_interns.assign(Score=scores).sort_values(by='Score', ascending=False)

    if len(sorted_interns) > 0:
        return sorted_interns.iloc[0]


```

```python
best_intern = hire_best_intern(df)
```
